```python
import pandas as pd
def load_data(file_path: str) -> pd.DataFrame:
    """
    Load data from a CSV file into a pandas DataFrame.

    Parameters:
    file_path (str): The path to the CSV file.

    Returns:
    pd.DataFrame: A DataFrame containing the loaded data.
    """
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"An error occurred while loading the data: {e}")
        return pd.DataFrame()  # Return an empty DataFrame on error
```

```python
Ob_data=load_data("BTC_1min.csv")
Ob_data
```

```python
trading_data=load_data("BTCUSD_1m_Combined_Index.csv")
trading_data
```

```python
trading_data['Open time'] = pd.to_datetime(trading_data['Open time'])
Ob_data['system_time'] = pd.to_datetime(Ob_data['system_time'])
Ob_data['system_time'] = Ob_data['system_time'].dt.tz_localize(None)

```

```python
start_time = Ob_data['system_time'].min()
end_time = Ob_data['system_time'].max()
filtered_trading_data = trading_data[
    (trading_data['Open time'] >= start_time) & 
    (trading_data['Open time'] <= end_time)
]
```

```python
print (Ob_data)
print(Ob_data.columns)
print(filtered_trading_data)
print(filtered_trading_data.columns)
```

```python
# If Close>Open : Buy signal 
# If Close<Open : Sell Signal
#
import numpy as np

# Step 1: Floor timestamps to the nearest minute
Ob_data = Ob_data.copy()
Ob_data['minute'] = Ob_data['system_time'].dt.floor('T')

filtered_trading_data = filtered_trading_data.copy()
filtered_trading_data['minute'] = filtered_trading_data['Open time'].dt.floor('T')

# Step 2: Compute midpoint per minute (mean if multiple OB snapshots per minute)
minute_midpoints = Ob_data.groupby('minute')['midpoint'].mean().sort_index()

# Step 3: Compute midpoint drift
midpoint_drift = minute_midpoints.diff()

# Step 4: Join midpoint drift into trading data
filtered_trading_data = filtered_trading_data.merge(
    midpoint_drift.rename('midpoint_drift'), on='minute', how='left'
)

# Step 5: Apply tick rule + midpoint drift to infer trade side
def infer_trade_side(row):
    tick_rule = 'BUY' if row['Close'] > row['Open'] else 'SELL' if row['Close'] < row['Open'] else 'UNKNOWN'
    midpoint_signal = 'BUY' if row['midpoint_drift'] > 0 else 'SELL' if row['midpoint_drift'] < 0 else 'UNKNOWN'

    # Majority vote
    if tick_rule == midpoint_signal:
        return tick_rule
    elif tick_rule != 'UNKNOWN':
        return tick_rule
    elif midpoint_signal != 'UNKNOWN':
        return midpoint_signal
    else:
        return 'UNKNOWN'

# Step 6: Create trade_side column
filtered_trading_data['trade_side'] = filtered_trading_data.apply(infer_trade_side, axis=1)
```
